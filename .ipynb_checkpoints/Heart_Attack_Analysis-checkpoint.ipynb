# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

# Dataset and attributes description

Dataset from https://archive.ics.uci.edu/ml/datasets/Heart+Disease

Variable definitions in the dataset:
- age: age of the patient
- sex: sex of the patient
- cp: chest pain type
 - 1 = typical angina
 - 2 = atypical angina
 - 3 = non-anginal pain
 - 4 = asymptomatic
- trestbps: resting blood pressure (in mm Hg on admission to the hospital)
- chol:serum cholestoral (in mg/dl)
- fbs: fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
- restecg: resting electrocardiographic results
 - 0 = normal
 - 1 = having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
 - 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria
- thalach: maximum heart rate achieved
- exang: exercise induced angina (1 = yes; 0 = no)
- oldpeak: ST depression induced by exercise relative to rest
- slope: the slope of the peak exercise ST segment
 - 1 = upsloping
 - 2 = flat
 - 3 = downsloping
- ca: number of major vessels (0-3) colored by flourosopy
- thal: displays the thalassemia
 - 3 = normal
 - 6 = fixed defect
 - 7 = reversable defect
- output: chance of a heart attack
 - 0 = less 
 - 1 = more


# Data loading and analysis

```python
csv_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak',
             'slope', 'ca', 'thal', 'output']

df = pd.read_csv('heart_attack.data', names=csv_names)
df.head()
```

```python
df['output'] = df['output'].replace([1,2,3,4], [1,1,1,1])
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

Analysis:
- data has 303 rows and 14 columns
- all the variables are in the numerical format (integer or float)
- there is no NaN values, but there are some '?' values to replace

# Missing value analysis

```python
tmp_df = pd.DataFrame(columns=csv_names)

for i in csv_names:
    tmp_df = pd.concat([tmp_df, df.loc[df[i].isin(['?'])]])    
    
tmp_df.sort_index()
```

There are some '?' in ca and thal column. It will be replaced with most popular wariable in these columns

```python
df['ca'] = df['ca'].replace(['?'],df['ca'].mode())
df['ca'].value_counts()
```

```python
df['thal'] = df['thal'].replace(['?'],df['thal'].mode())
df['thal'].value_counts()
```

Also there is need to change thal values from 3, 6, 7 to range 0-2.

```python
df['thal'].replace(['3.0', '7.0', '6.0'], ['0.0', '1.0', '2.0']).value_counts()
```

```python
df['thal'] = df['thal'].astype(str).astype(float)
df['ca'] = df['ca'].astype(str).astype(float)
```

```python
df.info()
```

And now every value is a float.

# Numerical variables analysis

b) analysis with pairplot </br>
c) statistics </br>
d) correlations

a) analysis with distplot </br>

```python
numerical_vars
```

```python
numerical_axis_names = ['Age of the patient', 'Resting blood pressure', 'Cholesterol',
                        'Maximum heart rate achieved', 'ST depression']
list(zip(numerical_vars, numerical_axis_names))
```

```python
title_font = {'family': 'arial', 'color': 'darkred', 'weight': 'bold', 'size': 15}
axis_font = {'family': 'arial', 'color': 'darkblue', 'weight': 'bold', 'size': 13}

x, y = plt.subplots(figsize=(8, 6))
sns.distplot(df['age'], hist=False, ax=y)
y.axvline(df['age'].mean(), color='r', ls='--')
y.axvline(df['age'].median(), color='g', ls='--')

plt.title('Age', fontdict=title_font)
plt.xlabel('Age of the patient', fontdict=axis_font)
plt.ylabel('Density', fontdict=axis_font)

print(df['age'].describe())
df['age'].mode().iat[0]
```

<b>Analysis of 'age' variable</b>
- the mode is 58 and mean is 54.4,
- so there is a little negative skew,
- the quartiles seems to be homogeneously distributed,
- so the data tends to be normally distributet with a slight left skew.

```python
x, y = plt.subplots(figsize=(8, 6))
sns.distplot(df['trestbps'], hist=False, ax=y)
y.axvline(df['trestbps'].mean(), color='r', ls='--')
y.axvline(df['trestbps'].median(), color='g', ls='--')

plt.title('Trestbps', fontdict=title_font)
plt.xlabel('Resting blood pressure', fontdict=axis_font)
plt.ylabel('Density', fontdict=axis_font)

print(df['trestbps'].describe())
df['trestbps'].mode().iat[0]
```

<b>Analysis of 'trestbps' variable</b>
- the mode is 120 and mean is 131.7,
- so there is a little positive skew,
- when we look at quartiles there is a big gap between 75% and max value,
- this shows that there is right skew in the data.

```python
x, y = plt.subplots(figsize=(8, 6))
sns.distplot(df['chol'], hist=False, ax=y)
y.axvline(df['chol'].mean(), color='r', ls='--')
y.axvline(df['chol'].median(), color='g', ls='--')

plt.title('Chol', fontdict=title_font)
plt.xlabel('Cholesterol', fontdict=axis_font)
plt.ylabel('Density', fontdict=axis_font)

print(df['chol'].describe())
df['chol'].mode().iat[0]
```

```python
x, y = plt.subplots(figsize=(8, 6))
sns.distplot(df['thalach'], hist=False, ax=y)
y.axvline(df['thalach'].mean(), color='r', ls='--')
y.axvline(df['thalach'].median(), color='g', ls='--')

plt.title('thalach', fontdict=title_font)
plt.xlabel('Maximum heart rate achieved', fontdict=axis_font)
plt.ylabel('Density', fontdict=axis_font)

print(df['thalach'].describe())
df['thalach'].mode().iat[0]
```

```python
x, y = plt.subplots(figsize=(8, 6))
sns.distplot(df['oldpeak'], hist=False, ax=y)
y.axvline(df['oldpeak'].mean(), color='r', ls='--')
y.axvline(df['oldpeak'].median(), color='g', ls='--')

plt.title('oldpeak', fontdict=title_font)
plt.xlabel('ST depression', fontdict=axis_font)
plt.ylabel('Density', fontdict=axis_font)

print(df['oldpeak'].describe())
df['oldpeak'].mode().iat[0]
```

```python
for i, z in list(zip(numerical_vars, numerical_axis_names)):
    plt.figure(figsize=(8,6))
    sns.distplot(df[i], hist_kws = dict(linewidth = 1, edgecolor = "k"))
    
    plt.title(i, fontdict=title_font)
    plt.xlabel(z, fontdict=axis_font)
    plt.ylabel("Density", fontdict=axis_font)
    
    plt.tight_layout()
    plt.show()
```
